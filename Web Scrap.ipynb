## Book Example 

```python
# goal: get title of every book with a 2 star rating
```

```python
base_url = 'http://books.toscrape.com/catalogue/page-{}.html'
```

```python
base_url.format('20')
```

```python
page_num = 12
base_url.format(page_num)
```

```python
res_book = requests.get(base_url.format(1))
```

```python
soup_book = bs4.BeautifulSoup(res_book .text,"lxml")
```

```python
soup_book
```

```python
products = soup_book.select('.product_pod')
```

```python
example = products[0]
```

```python
example.select('.star-rating.Three')
```

```python
example.select('.star-rating.Two')
```

```python
example.select('a')[1]['title']
```

```python
#we can check if sth is 2 stars(string call in, example.select(raing))
#example.select('a')[1]['title'] to grab book title
```

```python
two_star_titles = []

for n in range(1,51):
    scrape_url_res = requests.get(base_url.format(n))
    asoup = bs4.BeautifulSoup(scrape_url_res.text, 'lxml')
    books = asoup.select('.product_pod')
    
    for book in books:
        if len(book.select('.star-rating.Two')) != 0:
            book_title = book.select('a')[1]['title']
            two_star_titles.append(book_title)
            
```

```python
two_star_titles
```
